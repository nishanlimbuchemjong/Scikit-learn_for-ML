# Simple Imputer

```python
from sklearn.impute import SimpleImputer
```

```python
# we can fill missing values using mean
simple = SimpleImputer(strategy='mean')
df['A'] = simple.fit_transform(df[['A']])
df
```

```python
# we can also fill using median
simple = SimpleImputer(strategy='median')
df['B'] = simple.fit_transform(df[['B']])
df
```

```python
# And also we can fill using mode
simple = SimpleImputer(strategy='most_frequent')
df['C'] = simple.fit_transform(df[['C']])
df
```

```python
# And finally we can also fill the missing value using fillna()
df['D'] = df['D'].fillna(0)
df
```
